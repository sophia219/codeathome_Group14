Beschreibung der beiden Dateien und der jeweiligen Attribute:

**a) TripAdvisor_site.csv**
  
  1. **User_name**: Name des Benutzers, kann als Primärschlüsselspalte zum Join der beiden Tabellen benutzt werden

  2. **Attraction_name***: Name der bewerteten Attraktion

  3. **overall_rating***: Rating der Attraktion insgesamt (0-5)

  4. **Site_reviews**: Anzahl der Bewertungen für Attraktion => Indikator für Bekanntheit?

  5. **attraction_type**: TrypAdvisor Attraction-Kategorie für bewertete Attraktion (Museum/Park/etc.)

  6. **User_city**: Stadt des Benutzers

  7. **User_country**: Land des Benutzers

  8. **User_rating**: Rating der Attraktion des einzelnen Benutzers

  9. **text***: Text (mit max Zeichenanzahl) der Bewertung

  10. **U_review_count***: Anzahl der Bewertungen, die der Benutzer auf Tripadvisor verfasst hat

  11. **User_helpful***: Anzahl der Hilfreich-Wertungen für die Bewertung des Benutzers

  12. **year_visited**: Jahr des Besuchs der Attraktion (nicht des Verfassens der Bewertung!)

  13. **month_visited**: Monat des Besuchs der Attraktion (nicht des Verfassens der Bewertung!)

  ?14. **Travel_type**: Reiseart (als Paar/allein/mit Familie) wird noch hinzugefügt, falls ich das technisch hinbekomme (die Info ist etwas im html Text versteckt)


**b) TripAdvisor_user.csv**

  1. **User_name**: Name des Benutzers, kann als Primärschlüsselspalte zum Join der beiden Tabellen benutzt werden

2. **visited_attraction***: Name von weiterer besuchten Attraktion/Hotel/Restaurant eines Benutzers, der für die Attraktion in TripAdvisor_site.csv eine Bewertung geschrieben hat (bis zu 20 weitere Reviews des Benutzers werden gescrapt => mehr/weniger crawlen?)

3. **city**: Ort der besuchten Attraktion/etc. 

4. **country**: Land der besuchten Attraktion/etc.

5. **year_visited**: Jahr, in dem die Attraktion/etc. besucht wurde

6. **month_visited**: Monat, in dem die Attraktion/etc. besucht wurde

7. **overall_rating***: Rating der Attraktion/etc. insgesamt

8. **U_rating**: Rating der Attraktion/etc. des einzelnen Benutzers

9. **Site_review_counts**: Anzahl der Bewertungen für Attraktion/etc. => Indikator für Bekanntheit?

10. **category**: TripAdvisors Kategorie für Attraktion/etc. (Restaurant/Hotel/Atttraktion/Kreuzfahrt)

11. **attraction_type**: TrypAdvisor Attraction-Kategorie für bewertete Attraktion (Museum/Park/etc.)

12. **price**: Preiskategorie des Restaurants (€ bis €€€€) bzw. absoltuer Wert für Hotels von booking.com für eine Nacht (kann vielleicht auch in kategorische Variable umgewandelt werden?)
