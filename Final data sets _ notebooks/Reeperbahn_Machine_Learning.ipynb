```python
Ree = pd.read_csv('Reeperbahn_Reviews.csv')
Ree.rename(columns = {'profilename':'User_name'}, inplace = True)
Users = pd.read_csv('Reeperbahn_users.csv')
```

```python
# Remove Users with less than 5 reviews
users_to_drop = list()
for user,count in Users.groupby('User_name').count().Site_reviews_count.items():
    if count < 5:
        users_to_drop.append(user)
Users=Users[~Users.User_name.isin(users_to_drop)]
Ree = Ree[~Ree.User_name.isin(users_to_drop)]

def cleanReviewsCount(count):
    decimal_places = str(count).split('.')[1]
    if int(decimal_places) != 0:
        count = count*1000
    else: count = count
    return count

Users.loc[:,'Site_reviews_count']=Users.Site_reviews_count.apply(cleanReviewsCount)

# clean Ranking column
def cleanRanking(Ranking):
    if Ranking is not np.nan:
        counter =  0
        Ranking = Ranking.split(',')
        for i in Ranking:
            i = i.replace("'",'')
            i = i.replace('[','')
            i = i.replace(']','')
            i = i.replace(' ','')
            try: i = int(i)
            except: pass
            Ranking[counter] = i
            counter = counter + 1
    else: Ranking = Ranking
    return Ranking

Users.loc[:,'Ranking']=Users.Ranking.apply(cleanRanking)

# drop Reeperbahn reviews in User
Users = Users[Users.visited_attraction != 'Reeperbahn']
Users.reset_index(drop=True, inplace=True)
```

```python
# get an overview over all attraction types there are
all_att_types = list()
att_list = Users.attraction_type.unique()
for several_types in att_list:
    if several_types is not np.nan:
        several_types = several_types.split(',')
        for single_type in several_types:
            single_type = single_type.replace(' ','')
            if single_type not in all_att_types:
                all_att_types.append(single_type)
for each in all_att_types:
    print(each)
```

```python
# transfer attraction types into dummy variables
att_type_1 = list()
att_type_2 = list()
att_type_3 = list()
    
def makeList(types):
    counter = 0
    try:
        types = types.split(',')
        for each in types:
            each = each.replace(' ','')
            types[counter] = each
            counter = counter + 1
    except:
        types = np.nan
    try:
        att_type_1.append(types[0])
    except:
        att_type_1.append('none')
    try:
        att_type_2.append(types[1])
    except:
        att_type_2.append('none')
    try:
        att_type_3.append(types[2])
    except:
        att_type_3.append('none')
    print(types)
    return types

Users.loc[:,'attraction_type']=Users.attraction_type.apply(makeList)

type_dict = {'att_type_1':att_type_1, 'att_type_2':att_type_2, 'att_type_3':att_type_3}
dum = pd.DataFrame(type_dict)

dummies=dum.stack().str.get_dummies().sum(level=0)
dummies.drop(['none'],axis=1,inplace =True)
dummies
```

```python
# concat dummies with original DataFrame
df = pd.concat([Users,dummies], axis=1)
df
```

```python
# categorize price data
import sys
counter = 0
delete =  [np.nan, '€', '€€ - €€€', '€€€€']
price_list = df.price
for el in price_list:
    if el is not np.nan:
        el = el.replace('\xa0$','')
    price_list[counter] = el
    counter = counter + 1
        
counter = 0
for el in price_list:
    if (el != '€') and (el != '€€ - €€€') and (el != '€€€€') and (el is not np.nan):
        try:
            el = int(el)
        except: 
            el = int(float(el)*1000)
    price_list[counter] = el
    counter = counter + 1
len(price_list)

cat_price = list()
for el in price_list:
    if el is not np.nan:
        if el=='€':
            el = 0
        elif el == '€€ - €€€':
            el = 1
        elif el == '€€€€':
            el = 2
       
        else:
            if el < np.percentile([e for e in price_list if e not in delete], 33):
                el = 0
            elif el < np.percentile([e for e in price_list if e not in delete], 66):
                el = 1
            else: 
                el = 2
    else:
        el = 'none'
    cat_price.append(el)
df.loc[:,'price']=pd.Series(cat_price)
df
```

```python
# Based on the number of reviews of a site, a new variable is created
secret = None
def secret(reviews):
    if reviews < 50:
        secret = True
    else: secret = False
    return secret
df.loc[:,'secret']=df.Site_reviews_count.apply(secret)

popular = None
def popular(reviews):
    if reviews > 10000:
        popular = True
    else: popular = False
    return popular
df.loc[:,'popular']=df.Site_reviews_count.apply(popular)
```

```python
# drop not needed columns
final_df = df.drop(['visited_attraction', 'Ranking', 'city', 'country', 'Site_reviews_count', 'category', 'attraction_type'], axis = 1)
final_df
```

Approach to make a recommendation application, whether someone will like the Reeperbahn or not, based on his own preferences.

- not suitable in practice, since stated preferences may differ from characteristics of attractions, which someone actually visited

Better approach:
- ask the user of the application for attractions/hotels/restaurants he/she has visited over the last years

=> find out categories of attractions/hotels/restaurants on Tripadvisor in order to make the profile (not inthe extent of this project, Tripadvisor API would be probably necessary)


```python

profile = list()
def askQuestion(question):
    while True:
        attribute = input(question)
        if (attribute.lower().strip()== 'j'):
            profile.append(1)
            break
        elif (attribute.lower().strip()== 'n'):
            profile.append(0)
            break
        else: print('Input is not valid! Try again: ')
            
askQuestion('Do you like small, unknown attractions/hotels/restaurants? (j/n)')
askQuestion('Do you like big, well-known attractions/hotels/restaurants? (j/n)')
askQuestion('Are you an economic tourist? (j/n)')
askQuestion('Are you a luxury tourist (j/n)?')

while True:
    Reiseart = input('travel type: \n alone: 0 \n couple: 1 \n business: 2 \n friends: 3 \n familiy: 4 \n (type in number)')
    try: 
        Reiseart = int(Reiseart)
        if Reiseart == 0:
            profile.extend((1,0,0,0,0))
            break
        elif Reiseart == 1:
            profile.extend((0,1,0,0,0))
            break
        elif Reiseart == 2:
            profile.estend((0,0,1,0,0))
            break
        elif Reiseart == 3:
            profile.extend((0,0,0,1,0))
            break
        elif Reiseart == 4:
            profile.extend((0,0,0,0,1))
            break
        else: print('Input is not valid! Try again: ')
    except: print('Input is Not valid! Try again: ')

askQuestion('Do you like to explore urban districts? (j/n)')
askQuestion('Do you like history? (j/n)')
askQuestion('Do you like science? (j/n)')
askQuestion('Do like to relax? (j/n)')
askQuestion('Do you like art and architecture? (j/n)')
askQuestion('Do you like nature? (j/n)')
askQuestion('Do you like water (sea,lakes,etc)? (j/n)')
askQuestion('Do like thrill? (j/n)')
askQuestion('Do you like to do sport? (j/n)')
askQuestion('Do you like peace and quiet places? (j/n)')
askQuestion('Do you like shopping? (j/n)')
askQuestion('Do like culture? (j/n)')
askQuestion('Is good food on your vacation essential for you? (j/n)')

prediction = input_clf.predict(np.array(profile).reshape(1,-1))
if prediction[0]==0:
    print('You will most likely dislike the Reeperbahn!')
elif prediction[0]==1:
    print('You will most likely like the Reeperbahn!')
        
```
